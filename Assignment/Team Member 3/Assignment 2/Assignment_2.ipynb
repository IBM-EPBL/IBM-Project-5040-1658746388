# <span style='color:Green'>  Check Missing Values  </span>

```python
def num_obj_features(df):
    num_cols = []
    obj_cols = []
    s= df.dtypes
    for i in s.index:
        if (s[i] == 'int64' or s[i] == 'float64'):
            num_cols.append(i)
        elif s[i]== 'object':
            obj_cols.append(i)
        else:
            pass
    return num_cols, obj_cols

num_col, obj_col = num_obj_features(df)
```

```python
df[num_col].head()
df[obj_col].head()
```

# <span style='color:Green'>  Find And Replace Outliers  </span>

```python
 df.drop(['RowNumber', 'CustomerId', 'Surname'], axis = 1, inplace=True)
```

```python
def find_outlier(data):
    outlier = []
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    for i in data:
        z_score = (i - mean)/std
        if np.abs(z_score)>threshold:
            outlier.append(i)
    return outlier
Score_list = df['CreditScore'].tolist()
Balance_list = df['Balance'].tolist()
Salary_list = df['EstimatedSalary'].tolist()
```

```python
Score_outlier = find_outlier(Score_list)
Score_outlier
```

```python
Balance_outlier = find_outlier(Balance_list)
Balance_outlier
```

```python
Salary_outlier = find_outlier(Salary_list)
Salary_outlier
```
